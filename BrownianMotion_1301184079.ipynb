```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from statistics import mean
```

```python
def random_walk(x_min,y_min,x_max,y_max):
    xi = []
    yi = []
    
    
    x = random.randint(x_min,x_max)
    y = random.randint(y_min,y_max)
    
    x_range = x_max - x_min
    y_range = y_max - y_min
    xi.append(x)
    yi.append(y)
    
    n = 1000 #banyak iterasi
    
    for i in range(n):
        rand = random.uniform(0,1)
        if rand <= 0.25:
            x+=1
        elif (rand<= 0.5):
            y-=1
        elif (rand <=0.75):
            x-=1
        else:
            y+=1
        #update position
        xi.append(x)
        yi.append(y)
        
    return xi,yi
```

```python
x = []
y = []
distance_list = []

x_min,y_min = -25,-25
x_max,y_max = 25, 25


plt.figure(figsize=(10,10))

n_particle = 9
for i in range(n_particle):
    xi,yi = random_walk(x_min,y_min,x_max,y_max)
    x.append(xi)
    y.append(yi)
    distance = math.sqrt(xi[len(xi)-1]**2+yi[len(yi)-1]**2)
    distance_list.append(distance)
    
    #plot randomwalk particle
    plt.plot(x[i],y[i])
    
    plt.scatter(x[i][0],y[i][0], color = 'red') #titik awal
    plt.scatter(x[i][len(x[i])-1],y[i][len(y[i])-1], color = 'green') #titik akhir
    # titik awal dan akhir lalu di plot garis putus-putus
    plt.plot([x[i][0],x[i][len(x[i])-1]],[y[i][0],y[i][len(x[i])-1]], linestyle = '--')
    
    plt.xlabel("x")
    plt.ylabel("y")

print('Jarak maksimum = ',max(distance_list))
print('Jarak minimum = ',min(distance_list))
print('Jarak rata-rata = ',mean(distance_list))
```
